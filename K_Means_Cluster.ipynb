```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
link="https://raw.githubusercontent.com/codebasics/py/master/ML/13_kmeans/income.csv"
```

```python
df= pd.read_csv(link)
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["Age"],df["Income($)"],color="red",marker="*")
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3)
x=df[["Age", "Income($)"]]
df["cluster"]=km.fit_predict(x)
df
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1["Age"],df1["Income($)"],color="red")
plt.scatter(df2["Age"],df2["Income($)"],color="green")
plt.scatter(df3["Age"],df3["Income($)"],color="blue")
plt.xlabel("age")
plt.ylabel("income")
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
df["Income($)"]=mms.fit_transform(df[["Income($)"]])
df["Age"]=mms.fit_transform(df[["Age"]])
df
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3)
x=df[["Age", "Income($)"]]
df["cluster"]=km.fit_predict(x)
df
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
plt.scatter(df1["Age"],df1["Income($)"],color="red")
plt.scatter(df2["Age"],df2["Income($)"],color="green")
plt.scatter(df3["Age"],df3["Income($)"],color="blue")
plt.xlabel("age")
plt.ylabel("income")
```

```python
# sum of square
from sklearn.cluster import KMeans
sse=[]
for k in range(1,10):
  km=KMeans(n_clusters=k)
  km.fit(df[["Age","Income($)"]])
  sse.append(km.inertia_)
```
