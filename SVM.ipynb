```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
df1=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df2=pd.DataFrame(iris.target,columns=["Target"])
```

```python
df=pd.concat([df1,df2],axis=1)
df
```

```python
df[df.Target==0].count()
```

```python
df["flower_name"]=df.Target.apply(lambda x: iris.target_names[x])
df
```

```python
df_setosa=df[df.Target==0]
df_versicolor=df[df.Target==1]
df_virginica=df[df.Target==2]
```

```python
import matplotlib.pyplot as plt
```

```python
df_setosa['sepal length (cm)']
```

```python
plt.scatter(df_setosa['sepal length (cm)'],df_setosa['sepal width (cm)'],marker="*",color="red")
plt.scatter(df_versicolor['sepal length (cm)'],df_versicolor['sepal width (cm)'],marker="+",color="blue")
plt.scatter(df_virginica['sepal length (cm)'],df_virginica['sepal width (cm)'],marker="o",color="green")
```

```python
plt.scatter(df_setosa['petal length (cm)'],df_setosa['petal width (cm)'],marker="*",color="red")
plt.scatter(df_versicolor['petal length (cm)'],df_versicolor['petal width (cm)'],marker="+",color="blue")
plt.scatter(df_virginica['petal length (cm)'],df_virginica['petal width (cm)'],marker="o",color="green")
```

```python
x=df.iloc[:,:-2]
y=df["Target"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC(C=1,gamma=1)
model.fit(x_train,y_train)
```

```python
x_train
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[6.9,5.0,5.4,2.8]])
```
